```python
import pandas as pd
import numpy as np
import smote_variants as sv
```

```python
csv_file = "./raw/raw_cutoff_trainset.csv"
```

```python
df = pd.read_csv(csv_file)
df
```

```python
features = df.columns.to_list()
features
```

```python
X_train = df.drop('label', axis=1)
y_train = df['label']
```

```python
del df
```

```python
import gc
gc.collect()
```

```python
y_train.shape
```

```python
oversampling_methods = ['ADASYN'] # 'NRAS', 'SMOTE', 'Gaussian_SMOTE', 'kmeans_SMOTE','SYMPROD','CCR', 'SOMO', 
```

```python
for ovsampl in oversampling_methods:
    try:
        oversampler = sv.MulticlassOversampling(ovsampl)
        X_samp, y_samp = oversampler.sample(X_train, y_train)
        
        new_df = pd.DataFrame(X_samp, columns = features[:-1])
        new_df['label'] = y_samp 
        
        new_df.to_csv(f"{ovsampl}_trainset.csv")
        del new_df, X_samp, y_samp
        
    except Exception as e:
        print("====> ", e)
    
    gc.collect()
```
